```python
class individual_effect():
    def __init__(self, v, posx, posy, movex, movey, i, imcontagious, still): #defining variables of the class, there were more here ;)
        #speed at which a 'person' moves-dictates the number of interations of each person
        self.speed = v 
        
        #position of the person
        self.posx = posx
        self.posy = posy
        
        #where they will move too during the day
        self.movex = movex
        self.movey = movey
        
        #state of the person
        self.infected = False
        self.susceptible = True
        self.removed = False #'recovered' replaced as removed bc either immune or dead
        
        #specific person
        self.indice = i
        self.theperson = str(i)
        
        #quarantined// are they moving?
        self.still = still
        
        #movement per day
        if self.still:
            self.changex = 0
            self.changey = 0
        else:
            self.changex = (self.movex - self.posx) / self.speed
            self.changey = (self.movey - self.posy) / self.speed
        
        #when person is infected
        self.INFECT = -1
        
        #how long they stay infected
        self.imcontagious = imcontagious
        
    def __str__(self):
        return self.theperson + str(self.posx) + str(self.posy)
    
    def infecting_yous(self, i): #infecting these lads
        self.infected = True
        self.susceptible = False
        self.removed = False
        self.imcontagious = i
    
    def removing_yous(self, i): #removing these lads
        self.infected = False
        self.susceptible = False
        self.removed = True
    
    def next_move(self, movex, movey): #the next days place to move to, check this function is necessary
        self.movex = movex
        self.movey = movey
        if self.still: #would be still if quarantining
            self.changex = 0
            self.changey = 0
        else:
            self.changex = (self.movex - self.posx) / self.speed
            self.changey = (self.movey - self.posy) / self.speed
        print(self.movex, self.movey, self.indice) #not fully right
    
    def do_i_need_to_remove_you (self, i): #getting rid of people after their infection passes
        if self.INFECT>-1:
            if i - self.INFECT > self.imcontagious:
                self.removing_yous()
    
    def animation_station(self, new_posx, new_posy ): #animate these lads
        if new_posx==0 and new_posy==0:
            self.posx = self.posx + self.changex
            self.posy = self.posy + self.changey
        else:
            self.posx = new_posx
            self.posy = new_posy
        if abs(self.posx-self.movex)<3 and abs(self.posy-self.movey)<3:
            self.next_move(np.random.random()*100, np.random.random()*100) #limiting to a 100 x 100 box
        if self.posx>100:
            self.posx=100
        if self.posy>100:
            self.posy=100
        if self.posx<0:
            self.posx=0
        if self.posy<0:
            self.posy=0
        
    def colours_of_life(self):
        if self.infected:
            return 'red'
        elif self.susceptible:
            return 'green'
        elif self.removed:
            return 'blue'
            
    def people_places(self):
        return(self.posx, self.posy)
    
    def people_spaces(self, x, y): #distance between people
        return math.sqrt((self.posx-x)**2+(self.posy-y)**2)
    
```
